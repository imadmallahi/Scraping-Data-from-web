```python
import xml.etree.ElementTree as ET
import numpy as np
from scipy import sparse
from fast_pagerank import pagerank
from collections import Counter as C
import pandas as pn
```

# PageRank Implementation 

```python
root = ET.parse("graph.xml").getroot()

names = []
print('Les noeuds : ')
for node in root:
    print(node.find('name').text)
    names.append(node.find("name").text)
  
    
    
```

```python
    
dic={}


for n in names:
    
    dic[n]=names.index(n)
    

dic
```

```python
matrix=[]
for n,r in zip(names,root):
    for  link in r.findall("link"):
        matrix.append([dic[n],dic[link.text]])       
        
   
matrix = np.matrix(matrix)
print('la matrice des liaisons :')
matrix
```

```python
vector = np.asarray(matrix[:,0])
v=[]
for i in range(len(vector)):
    v.append(matrix[i,0])
    


dim = len(names)
#crÃ©e les matrices D T de dimention de nb des noeuds
T = np.zeros((dim,dim))
D=T
T


```

```python
for i in range(len(matrix)):
    for j in range(1):        
        j1=matrix[i,1]
        i1=matrix[i,j]
        #insertion ds valeur 
        T[i1][j1]= 1/C(v).get(i1)
        
lamb = 0.85
T = lamb * T
T

for i in range(len(T)):
    if(np.sum(T[i])!=0):
        D[i]=(1-lamb)/dim
    else :
        D[i]=1/dim

D

P = T+D
print(T,'\n + \n' ,D ,' \n =\n', P)

```

```python
R = [ 1/dim for k in range(dim)]

print('R',R)
for i in range(1000):
    Rn=np.tensordot(R, P, axes=1)
    print('Rn (',i,') :',Rn) 
    if (sum(np.abs(Rn-R))==0): 
        break
        
    else :
        R=Rn
        
        
```

```python
print('R (',i,') => :',R * 100)    
```
