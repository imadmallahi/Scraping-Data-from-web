```python
import os
from prettyprinter import pprint
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
from nltk.stem.porter import PorterStemmer
from gensim import corpora,similarities
import pandas as pd

number=[]
titles=[]
nars=[]
files=[]

def remove_html_tags(text):
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

path = 'PATopics2011/PATopics2011/'

fil= []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.txt' in file:
            files.append(os.path.join(r, file))

for f in files:
    re = open(f, "r")
    text=re.read()
    T=remove_html_tags(text)
    T=T.splitlines(False)
    
    number.append(T[1])
    titles.append(T[2])
    fil.append(T[4])

data2 = {'number':number,
         'files':fil,
        'title':titles
       }

# Create DataFrame
Total = pd.DataFrame(data2)

tokenized_documents= [[token for token in document.lower().split()] for document in titles]
#pprint(tokenized_documents)
stopwords_en = stopwords.words('english')
#print(stopwords_en)
TT=[]
for y in tokenized_documents:
    tokens_clean = [t for t in y if t not in stopwords_en]
    #pprint(tokens_clean)
    TT.append(tokens_clean)

pprint(TT)
    


```

```python
dictionary = corpora.Dictionary(TT)
print(dictionary)
pprint(dictionary.num_docs)
```

```python
print(dictionary.token2id)
print("\n",dictionary)
```

```python
corpus_doc2bow_vectors=[dictionary.doc2bow(tok_doc) for tok_doc in tokenized_documents]

for c in corpus_doc2bow_vectors:
    print(c)
```

```python
from gensim.models import TfidfModel
from gensim.corpora import Dictionary


%time  
tfidf_model = TfidfModel(corpus_doc2bow_vectors,id2word=dictionary,normalize=False)
corpus_tfidf_vector=tfidf_model[corpus_doc2bow_vectors]

print("******TFIDF_vector of the three documents ******")
for doc in corpus_tfidf_vector:
    print(doc)
```
